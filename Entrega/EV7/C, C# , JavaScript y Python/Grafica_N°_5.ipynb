```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/C, C# , JavaScript y Python.csv')

data['Semana'] = pd.to_datetime(data['Semana'], format='%Y-%m-%d')

plt.figure(figsize=(12, 6))
plt.plot(data['Semana'], data['C#'], marker='o', label='C#', color='red')
plt.plot(data['Semana'], data['C'], marker='o', label='C', color='blue')
plt.plot(data['Semana'], data['Python'], marker='o', label='Python', color='green')
plt.plot(data['Semana'], data['JavaScript'], marker='o', label='JavaScript', color='orange')

plt.title('Comparacin de Popularidad de Lenguajes de Programacin: C#, C, Python y JavaScript')
plt.xlabel('Semana')
plt.ylabel('Porcentaje de Bsqueda')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
