```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Backend_Frameworks.csv')

data['Semana'] = pd.to_datetime(data['Semana'], format='%Y-%m-%d')

plt.figure(figsize=(12, 6))

plt.plot(data['Semana'], data['Laravel'], marker='o', label='Laravel', color='blue')
plt.plot(data['Semana'], data['Django'], marker='o', label='Django', color='red')
plt.plot(data['Semana'], data['Spring Boot'], marker='o', label='Spring Boot', color='green')
plt.plot(data['Semana'], data['Express.js'], marker='o', label='Express.js', color='purple')
plt.plot(data['Semana'], data['ASP.NET Core'], marker='o', label='ASP.NET Core', color='orange')

plt.title('Anlisis de la Popularidad de Frameworks Backend: Laravel, Django, Spring Boot, Express.js y ASP.NET Core')
plt.xlabel('Tiempo')
plt.ylabel('Porcentaje de Bsqueda')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
