```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Lenguajes_Programacion_Orientados_a_Objetos.csv')

data['Semana'] = pd.to_datetime(data['Semana'], format='%Y-%m-%d')

plt.figure(figsize=(12, 6))
plt.plot(data['Semana'], data['C++'], marker='o', label='C++', color='blue')
plt.plot(data['Semana'], data['C#'], marker='o', label='C#', color='red')
plt.plot(data['Semana'], data['Java'], marker='o', label='Java', color='green')
plt.plot(data['Semana'], data['Kotlin'], marker='o', label='Kotlin', color='orange')

# Configuracin del ttulo y etiquetas
plt.title('Comparacin de Popularidad de Lenguajes de Programacin: C++, C#, Java y Kotlin')
plt.xlabel('Semana')
plt.ylabel('Porcentaje de Bsqueda')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

# Mostrar la grfica
plt.tight_layout()
plt.show()
```
