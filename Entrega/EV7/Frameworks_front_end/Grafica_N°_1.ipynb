```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/Frameworks_front_end.csv')

data['Semana'] = pd.to_datetime(data['Semana'], errors='coerce')

plt.figure(figsize=(12, 6))

plt.plot(data['Semana'], data['React'], marker='o', label='React', color='blue')
plt.plot(data['Semana'], data['Angular'], marker='o', label='Angular', color='red')
plt.plot(data['Semana'], data['Vue.js'], marker='o', label='Vue.js', color='green')

plt.title('Anlisis de la Popularidad de Frameworks Frontend: React, Angular y Vue.js')
plt.xlabel('Tiempo')
plt.ylabel('Porcentaje de Busqueda')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
