```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/PostgreSQL, MySQL, SQLite, MongoDB.csv')

data['Semana'] = pd.to_datetime(data['Semana'], format='%Y-%m-%d')

plt.figure(figsize=(12, 6))

plt.plot(data['Semana'], data['PostgreSQL'], marker='o', label='PostgreSQL', color='blue')
plt.plot(data['Semana'], data['MySQL'], marker='o', label='MySQL', color='red')
plt.plot(data['Semana'], data['SQLite'], marker='o', label='SQLite', color='green')
plt.plot(data['Semana'], data['MongoDB'], marker='o', label='MongoDB', color='orange')

plt.title('Comparacin de Popularidad de Bases de Datos: PostgreSQL, MySQL, SQLite y MongoDB')
plt.xlabel('Semana')
plt.ylabel('Porcentaje de Bsqueda')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
