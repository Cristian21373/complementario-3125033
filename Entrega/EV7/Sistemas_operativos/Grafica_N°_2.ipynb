```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('/content/Sistemas_operativos.csv')

data['Semana'] = pd.to_datetime(data['Semana'], format='%Y-%m-%d')

plt.figure(figsize=(12, 6))

plt.plot(data['Semana'], data['GNU/Linux'], marker='o', label='GNU/Linux', color='blue')
plt.plot(data['Semana'], data['Microsoft Windows'], marker='o', label='Microsoft Windows', color='red')
plt.plot(data['Semana'], data['macOS'], marker='o', label='macOS', color='green')

plt.title('Anlisis de la Popularidad de Sistemas Operativos: GNU/Linux, Windows y macOS')
plt.xlabel('Tiempo')
plt.ylabel('Porcentaje de Uso')
plt.xticks(rotation=45)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
