```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Python, R, SAS , MATLAB.csv')

df["Semana"] = pd.to_datetime(df["Semana"])

plt.figure(figsize=(12, 6))
plt.plot(df["Semana"], df["R"], marker="o", label="R", color="blue")
plt.plot(df["Semana"], df["SAS"], marker="o", label="SAS", color="orange")
plt.plot(df["Semana"], df["MATLAB"], marker="o", label="MATLAB", color="green")
plt.plot(df["Semana"], df["Python"], marker="o", label="Python", color="red")

plt.title("Comparacin de Popularidad: R, SAS, MATLAB y Python")
plt.xlabel("Semana")
plt.ylabel("Porcentaje de Bsqueda")
plt.xticks(rotation=45)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
